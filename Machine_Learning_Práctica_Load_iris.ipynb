# HIPERPARAMETROS

```python
scores = []
for i in range (1, 51): #bucle 
    model = KNeighborsClassifier (n_neighbors=i) #i: que es el numero que vamos a ir variando
    model.fit(x_train, y_train)
    pred = model.predict (x_test)
    scores.append (accuracy_score (y_test, pred))
    
plt.plot (range (1,51), scores)  #Podemos ver los valores y performace, varian valores: hay valores que son optimas 
#para metricas, el mejor valor se puede elegir en los que figuran mas arriba. 
```
