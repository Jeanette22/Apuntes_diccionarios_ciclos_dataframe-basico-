### **LIBRERIAS PARA LEER BASE DE DATOS SQL**

#### OBJETIVO: CREAR UNA CONEXION A UNA BASE DE DATOS SQL 

```python
#Librerias necesarias: 
import sqlite3 as sq3               
import pandas.io.sql as pds             
import pandas as pd
```

```python
path = '/content/drive/MyDrive/classic_rock.db'
con = sq3.Connection(path)
```

**VEO LAS TABLAS QUE TIENE LA BASE DE DATOS: seleccionando una de las tabla_en este caso rock_songs** 

```python
query = '''
SELECT * 
FROM rock_songs;
'''

# Execute the query
observations = pds.read_sql(query, con)

observations.head()
```

```python
query = '''
SELECT * 
FROM rock_plays;
'''

# Execute the query
observations2 = pds.read_sql(query, con)

observations2.head()
```

```python
observations2.info()
```

```python
observations2.describe()
```

```python
observations2.isnull()
```

```python
#Eliminar datos duplicados 
observations3=observations2.drop_duplicates ()
observations3
```

```python
observations3[observations3.duplicated()].shape
```

```python
#Ordeno los datos 
observations3.sort_values ('date_time', ascending= False)
```
